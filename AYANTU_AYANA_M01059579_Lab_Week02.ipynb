4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
# Ask user for weight in pounds and height in inches
weight_pounds = float(input("enter weight in pounds: "))
height_inches = float(input("enter height in inches: "))

# Convert to metric units
weight_kg = weight_pounds * 0.45359237
height_m = height_inches * 0.0254

# Calculate BMI
bmi = weight_kg / (height_m ** 2)
print("BMI is:", round(bmi, 2))  # Round to 2 decimal places

# Determine BMI category
if bmi < 18.5:
    print("Underweight")
elif bmi <= 25:
    print("Normal")
elif bmi <= 30:
    print("Overweight")
else:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
# List of student marks
marks = [45, 67, 80, 35, 50, 70, 90, 55]

# Calculate average using sum() and len()
avg = sum(marks) / len(marks)

# Find the highest and lowest marks
highest = max(marks)
lowest = min(marks)

# Count passed and failed students
passed = len([m for m in marks if m >= 50])
failed = len([m for m in marks if m < 50])

# Print results
print("Average:", avg)
print("Highest:", highest)
print("Lowest:", lowest)
print("Passed:", passed)
print("Failed:", failed)
````

15. **Find the highest score**

Write a program that prompts the user to enter the number of students and each student’s name and score and display the student name and highest score.


````python
# Ask how many students there are
number_students = int(input("Enter number of students: "))

# Create two empty lists to store names and scores
names = []
scores = []

# Get each student's name and score
for i in range(number_students):
    name = input("Enter student name: ")
    score = float(input("Enter student score: "))
    names.append(name)
    scores.append(score)

# Find the highest score
highest_score = max(scores)

# Find the student with the highest score
top_student = ""
for i in range(len(names)):
    if scores[i] == highest_score:
        top_student = names[i]
        break

# Display all students and scores
print("\nAll students and their scores:")
for i in range(number_students):
    print(names[i], "-", scores[i])

# Display the student with the highest score
print("\nStudent with the highest score:", top_student)
print("Highest score:", highest_score)
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
# Import the random module to generate a secret number
import random 

# Generate a secret number between 1 and 100
secret = random.randint(1, 100)  
tries = 0  # Initialize number of guesses
guess = 0  # Initialize guess

# Keep asking until the correct number is guessed
while guess != secret:
    guess = int(input("Guess the number between 1 and 100: "))  # Ask for user's guess
    tries += 1  # Increase the number of tries
    if guess < secret:
        print("Too small!")  # Tell user the guess is too low
    elif guess > secret:
        print("Too large!")  # Tell user the guess is too high

# When guessed correctly
print("Correct! You guessed it in", tries, "tries.")  # Show number of attempts
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
actual = input("Enter actual password: ") # Get actual password
guess = input("Enter guessed password: ") # Get guessed password

# Count matching characters
matches = 0
for i in range(len(actual)):
    if i < len(guess) and guess[i] == actual[i]:
        matches += 1

# Calculate accuracy in percent
accuracy = round((matches / len(actual)) * 100)

# Show result
if accuracy == 100:
    print("Password cracked!")
print("Accuracy:", accuracy, "%")
````
