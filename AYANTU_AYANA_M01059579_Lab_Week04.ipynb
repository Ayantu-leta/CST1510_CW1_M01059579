1. Ask user for an input of his/her password and store it in a variable called password
Print the password length first then create function validating if the password has more than eight characters. If the password fulfills the above condition and has more then eight character print 'valid password' otherwise print invalid password and ask the user to enter the password again.

````python
# AYANTU AYANA M01059579

#Password validation (uppercase, lowercase, digit, min 8 chars)
while True:
    password = input('enter password: ')
    upper = any (c.isupper() for c in password)  # True if any uppercase letter exists
    lower = any (c.islower() for c in password)  # True if any lowercase letter exists
    digit = any (c.isdigit() for c in password) # True if any digit exists

    if len(password)>= 8 and upper and lower and digit:
        print('valid password')
        break # Exit the loop
    else:
        print('invaild password try again.')  # Ask again if rules not met
  
````

3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
#AYANTU AYANA M01059579

#Convert password to ASCII numbers
def password_ascii(pwd):
    result = ''
    for c in pwd:
        result += str(ord(c)) # Convert each character to ASCII and add to string
    return result
password = input('enter password: ')
print(password_ascii(password))  # Display ASCII string
````

````python
#AYANTU AYANA M01059579

#Binary to decimal conversion
binary = input('enter binary: ')
decimal = 0
for digit in binary:
    decimal = decimal * 2 + int(digit)  # Shift left and add current bit
print(decimal)
````
