### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
#AYANTU AYANA M01059579

#record daily temperatures from user input
temp = []

for i in range (1, 6):
    t = float(input(f"day {i} temp:")) #ask user to enter temperature
    temp.append(t)  #store temperature in list

#write all temperatures to a file
with open ('temp.txt', 'w') as f:
    for i, t in enumerate(temp, 1):
        f.write(f"day {i}: {t}°c\n")
#calculate and print average temperature
avg = sum(temp) / len(temp)
print('average temp:', avg,'°c')
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
#AYANTU AYANA M01059579

#dictionary to store clothing items for a set of sizes
sizes = {}
with open('clothing_stock.txt') as f:  #open the file for reading
    for line in f:
        item, size = line.strip().split(',')  #split line into item and size
        sizes.setdefault(item, set()).add(size)  #add size to the item's set

#print number of different sizes for each item
for item, s in sizes.items():
    print(item, ':', len(s), 'sizes')
`````

### Activity 17: Counting Words in Les_Brown.txt

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
#AYANTU AYANA LETA

#read Les_Brown.txt and count the occurrences of each word
word_count = {}

with open("Les_Brown.txt", "r") as f:
    for line in f:
        words = line.strip().lower().split()   #split line into words and convert to lowercase
        for word in words:
            word_count[word] = word_count.get(word, 0) + 1  #count each word

#write word counts to a new file
with open("word_count.txt", "w") as f:
    for word, count in word_count.items():
        f.write(f"{word}: {count}\n")

# Print word counts
for word, count in word_count.items():
    print(word, ":", count)
`````

### Activity 18 (List): Removing Duplicates from Clothing Stock

Write a Python program that reads the `clothing_stock.txt` file, removes any duplicate entries of clothing items with the same size, and then writes the unique entries to a new file called `unique_clothing_stock.txt`.

**Steps:**
1. Read the clothing items into a list.
2. Remove duplicates by storing them in a set.
3. Write the unique entries to a new file called `unique_clothing_stock.txt`.

<hr>

`````python
#AYANTU AYANA M01059579

#read clothing_stock.txt and store unique items
unique_items = set()

with open('clothing_stock.txt', 'r') as file:
    for line in file:
        unique_items.add(line.strip())  #remove extra spaces/newlines and add item to the set

#write unique items to a new file called unique_clothing_stock.txt
with open('unique_clothing_stock.txt', 'w') as file:
    for item in unique_items:
        file.write(item + '\n')

#print the unique items 
for item in unique_items:
    print(item)

`````

### Activity 19 (Set): Splitting Clothing Stock by Size
